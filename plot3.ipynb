```python
from numpy.random import randn
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(randn(30).cumsum(),'ko--')
#ko-- is for the dots and dashes
```

```python
#or we can write it as 
plt.plot(randn(30).cumsum(),color='k',linestyle='dashed',marker='o')
```

```python
data = np.random.randn(30).cumsum()
```

```python
plt.plot(data,'k--',label='Default')
```

```python
plt.plot(data,'k-',drawstyle='steps-post',label='steps-post')
```

```python
plt.legend(loc='best')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())
```

```python
#tickets to place the tickets along the axis
tickets = ax.set_xticks([0,250,500,750,1000])
lables =  ax.set_xticklabels(['one','two','three','four','five'],rotation=30,fontsize='small')
ax.set_title('first matplotlib')
ax.set_xlabel('stages')
```

```python
fig = plt.figure();ax=fig.add_subplot(1,1,1)
ax.plot(randn(1000).cumsum(),'k',label='one')
ax.plot(randn(1000).cumsum(),'k--',label='two')
ax.plot(randn(1000).cumsum(),'k.',label='three')
ax.legend(loc='best')


```

```python
#to draw shapes
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
rect = plt.Rectangle((0.2,0.75),0.4,0.15,color='k',alpha=0.3)
circ = plt.Circle((0.7,0.2),0.15,color='b',alpha=0.3)
pgon = plt.Polygon([[0.15,0.15],[0.35,0.4],[0.2,0.6]],color='g',alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```
